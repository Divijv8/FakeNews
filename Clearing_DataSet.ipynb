```python
import re
import pandas as pd

def clean_text(text):
    if not isinstance(text, str):
        return text

    # Remove Reuters header like:
    # "WASHINGTON (Reuters) -", "(Reuters) -", "LONDON (Reuters) "
    text = re.sub(r'\b[A-Z\s]*\(?Reuters\)?\s*[-]?\s*', '', text)

    # Remove tweet headers like "RT @user:" or mentions
    text = re.sub(r'RT\s*@\w+:', '', text)
    text = re.sub(r'\(@\w+\)', '', text)

    # Remove URLs if present
    text = re.sub(r'http\S+|www\S+', '', text)

    # Clean extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    return text

# Apply cleaning
final_combined_df['text'] = final_combined_df['text'].apply(clean_text)

#  Remove rows that have only Reuters info (no actual content)
# e.g., text length < 30 or empty after cleaning
final_combined_df = final_combined_df[final_combined_df['text'].str.len() > 30]

# Reset index after dropping rows
final_combined_df = final_combined_df.reset_index(drop=True)

```
